# 2. 举例讲解自定义图标样式
## 2.1 两种绘图接口
  matplotlib绘图有两种接口来处理。一种
  1. 直接用pyplot函数来绘图
  2. 另一个是基于对象(oo)的绘图方法,通过绘图后返回的各种对象，通过操作对象的方法来绘图
  
接下来以oo接口来说明整个绘图周期

## 2.2 要用到的数据
 

````python
# sphinx_gallery_thumbnail_number = 10
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

data = {'Barton LLC': 109438.50,
        'Frami, Hills and Schmidt': 103569.59,
        'Fritsch, Russel and Anderson': 112214.71,
        'Jerde-Hilpert': 112591.43,
        'Keeling LLC': 100934.30,
        'Koepp Ltd': 103660.54,
        'Kulas Inc': 137351.96,
        'Trantow-Barrows': 123381.38,
        'White-Trantow': 135841.99,
        'Will LLC': 104437.60}
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)
````

## 2.3 将数据绘制为柱状图(barplot)
  因为使用oo接口，所以首先我们要创建一个figure和axes(子图)的对象，通过
  ```python
plt.subplots()
```
  不传入参数可以直接创建，返回值为一个figure和axes的列表，通过unpack方法解包，这样我们就得到了一个``有一个axes（子图）的figure``

````python
fig, ax = plt.subplots()
````

  因为有了axes对象了，所以直接可以在axes（子图）对象中绘制图表，方法是调用axes对象的各种绘图函数：

````python
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
````

## 2.4 调用效果模板（style）
  图表有很多默认的已经配置好的效果模板（style），直接调用就可以使用，要看可以改变的styles列表，可以查看`plt.style.available`

````python
print(plt.style.available)
````

知道有什么效果模板后，我们可以直接应用指定的style：
```python
plt.style.use('fivethirtyeight')
```
  具体应用后效果，重绘图表效果如下

````python
plt.style.use('Solarize_Light2')
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
plt.show()
````

## 2.5 自定义调整图表样式
### 2.5.1 调整坐标轴
通过调用`style`可以直接调整整个图标样式，但如果要细调图表，需要手动自定义调整。
如：如果要调整x轴的值为倾斜方便显示。

首先我们要取得x轴的值的对象，通过下面方法获取：
```python
axes.Axes.get_xticklabels()
```
获取到对象后，我们可以通过调用之前说的```pyplot.setp()```方法来设置对象的属性

如：
* 倾斜度
* 轴文本size
* ...
```python
plt.setp(labels, rotation=45, horizontalalignment='right',,fontsize=10)
```

````python
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right',fontsize=15)

plt.show()
````

### 2.5.2 为坐标轴添加标签
然后我们还需要为坐标轴添加轴名，因为添加坐标轴是为axes添加xlabel，ylabel所以调用`plt.set()`为axes对象添加属性：
```python
# 指定x轴的范围为[-10000,140000],和xlabel，ylabel,title
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
```

````python
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

# 指定x轴的范围为[-10000,140000],和xlabel，ylabel,title
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
plt.show()
````

### 2.5.3 自定义坐标轴的数字标签样式
我们还可以自定义坐标轴上的值的显示模式，然后将这个函数传入`ticker.FuncFormatter`方法构成一个formateter对象，如：

````python
def currency(x, pos):
    # 定义一个函数，参数x为值，pos为这个值在坐标轴显示的位置
    if x >= 1e6:
        s = '${:1.1f}M'.format(x*1e-6)
    else:
        s = '${:1.0f}K'.format(x*1e-3)
    # 返回值是字符串，也就是要显示的内容
    return s

formatter = FuncFormatter(currency)
````

获得formatter后，要给坐标轴设置上这个formatter，则通过`axes`对象的`ax.xaxis,ax.yaxis`来获得坐标轴对象。
然后调用坐标轴的
```python
ax.xaxis.set_major_formatter(formatter)
```
方法设置formatter

````python
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')

ax.xaxis.set_major_formatter(formatter)
plt.show()
````

#### 2.5.3.1 位标签启用科学计数法
通过调用axes对象的`ticklabel_format()`方法可以为x轴y轴设置科学计数法标签：
```python
ax.ticklabel_format(style='sci', axis='x', scilimits=(0,0),useMathText=True)
# style 指定风格为科学计数法
# axis 指定为x轴启用
# scilimits (m,n)为在10^m 到10^n次方范围内启用科学计数法，直接(0,0)为为所有数启用科学计数法
# useMathText 如为True，表示方法为10^x，如为False表示方法为10Ex
```

````python
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

# 指定x轴的范围为[-10000,140000],和xlabel，ylabel,title
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
ax.ticklabel_format(style='sci', axis='x', scilimits=(0,0),useMathText=True)
plt.show()
````

### 2.5.4 调整图表的尺寸
在用`pyplot.subplots() `创建子图的时候，可以传入`figsize=(w, h)`指定图标的尺寸，w为宽度h为高度

````python
fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')

plt.show()
````

### 2.5.5 调整坐标轴的刻度分布
`ax.set_xticks(),ax.set_yticks()`方法可以设置的坐标轴在哪些值上显示值标签
如，下面为在0,25e3等位置显示标签：
```python
ax.set_xticks([0, 25e3, 50e3, 75e3, 100e3, 125e3])
```

````python
fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
ax.set_xticks([0, 25e3, 50e3, 75e3, 100e3, 125e3])
ax.xaxis.set_major_formatter(formatter)
plt.show()
````

### 2.5.5 绘制多条曲线
在一个图表中绘制多个图线
方法很简单，就是直接调用axes对象的各种绘图方法，就能在同一个图表中绘制多个图线

````python
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

# 绘制一条垂直线
ax.axvline(group_mean, ls='--', color='r')

# 为图表右边对应位置添加文本注解
for group in [3, 5, 8]:
    ax.text(145000, group, "New Company", fontsize=10,
            verticalalignment="center")

# 让标题的位置稍微离开图表一点
ax.title.set(y=1.05)

# 设置图表的label等文本
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')

# 设置label的formatter
ax.xaxis.set_major_formatter(formatter)

# 设置要显示刻度的位置
ax.set_xticks([0, 25e3, 50e3, 75e3, 100e3, 125e3])
#fig.subplots_adjust(right=0.1)

plt.show()
````

# 3. 图表中文日文等的乱码的解决
使用中文的话会乱码，比如：

````python
plt.plot([1,2,3])
plt.title("这是个标题")
````

可以看到标题是方块字，原因是matplotlib默认的字体不支持中文，解决办法就是换个字体。

首先需要显示出来支持的字体，看最后。

选择一个字体，比如仿宋，`FangSong`.
首先生成一个Font对象，通过调用实现。
```python
matplotlib.font_manager.FontProperties()
```
然后绘制title时指定font对象。

````python
from matplotlib.font_manager import FontProperties

plt.plot([1,2,3])
# 生成一个font对象，注意字体大小得在这里指定
font = FontProperties("fangSong",size=30)
# 传入字体对象
plt.title("这是个标题",fontproperties=font)
````

## 3.2 列举所有可用的字体

````python
import matplotlib.font_manager
import pandas as pd
fonts = matplotlib.font_manager.findSystemFonts()
l = []
for f in fonts:
    try:
        font =matplotlib.font_manager.FontProperties(fname=f)
        #print(font.get_family())
        l.append((f, font.get_name(), font.get_family(), font.get_weight()))
    except:
        pass
pd.set_option("max_rows",1000)
df = pd.DataFrame(l, columns=['path', 'name', 'family', 'weight'])
df
````
