````python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
````

# 1 图例的使用
要开启图例，需要指定曲线的名字，通过绘制时传入label参数指定。

然后调用
```python
plt.legend()
```

````python
plt.plot([1,2,3], label='Line 2')
plt.plot([3,2,1], label='Line 1')
plt.legend()
````

也可以绘图时不指定曲线的名字，而在显示图例时手动指定，如：
```python
plt.plot([1,2,3], label='Line 2')
plt.plot([3,2,1], label='Line 1')
plt.legend(["Lable1","Label2"])
```

````python
plt.plot([1,2,3], label='Line 2')
plt.plot([3,2,1], label='Line 1')
plt.legend(["Lable1","Label2"])
````

如果要只显示指定曲线的图例，只需要手动传入曲线对象和对应的图例名称。

````python
line_up, = plt.plot([1,2,3], label='Line 2')
line_down, = plt.plot([3,2,1], label='Line 1')
plt.legend([ line_down], [ 'Line Down'])
````

# 2. 图例指定位置
给`legend()`传入指定参数:loc

loc支持传入相对位置，如左上右上，具体见[legend()](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html#matplotlib.pyplot.legend)说明文档的loc部分。

loc也支持支持传入元祖,为xy的值对。

````python
line_up, = plt.plot([1,2,3], label='Line 2')
line_down, = plt.plot([3,2,1], label='Line 1')
plt.legend(loc='center')
````

# 3. 手动添加图例
如果一个图里没有绘制曲线，那么就没有图例对象可以显示。

这时候可以自己创建一个图例的对象（handle）,图例对象实际上就是各种曲线对象，只要是能绘制的曲线对象，都能拿来当图例，如，plt.plot()是绘制一条Line2D的曲线，

````python
import matplotlib
blue_line = matplotlib.lines.Line2D([], [], color='blue', marker='*',
                          markersize=15, label='Blue stars')
plt.legend(handles=[blue_line])

plt.show()
````

# 4. 绘制分割位置的图例

利用上面的绘制单个图例的方法，结合下面的指定图例位置可以绘制多个位置分割的图例，如：

````python
line_up, = plt.plot([1,2,3], label='Line 2')
line_down, = plt.plot([3,2,1], label='Line 1')

# 生成一个legend对象
legend1 = plt.legend([ line_down], [ 'Line Down'],loc="upper right")
# 将legend对象添加到Axes
plt.gca().add_artist(legend1)
# 同上面
legend2 = plt.legend([line_up],['Line Up'],loc="center")
plt.gca().add_artist(legend2)
````
