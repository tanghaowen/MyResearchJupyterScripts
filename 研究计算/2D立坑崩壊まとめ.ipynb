```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
breakdown_mu = [5,8,12]
breakdown_coh  = [1e5,1e5,1e5]
nonbreakdown_mu = [5,5,5]
nonbreakdown_coh = [1.2e5,1.5e5,2.0e5]
```

```python
plt.figure(figsize=(4,4))
plt.scatter(breakdown_coh,breakdown_mu,c="r",marker="x",label="Fault")
plt.scatter(nonbreakdown_coh,nonbreakdown_mu, label="Non Fault")
plt.grid()
plt.axis([0,2.5e5,0,14])
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0),useMathText=True)
plt.xlabel("Well Cohesion $\mathrm{N/m^2}$")
plt.ylabel("Well Friction Coefficient")
plt.legend()

plt.savefig("崩壊状況.png",quality=100,dpi=200)
```

# 流入粒子量による壁面積の変化

```python
plt.figure(figsize=(4,4))
particle_num = [873,1802,2732,3660]
area_change = [7.69,12.69,105,302]
plt.scatter(particle_num,area_change)
plt.grid()
plt.axis([0,4000,0,350])
plt.xlabel("Particle Number")
plt.ylabel("Area Change $\mathrm{m^2}$")
plt.tight_layout()
plt.savefig("流入粒子量による壁面積の変化.png",quality=100,dpi=200)
```
